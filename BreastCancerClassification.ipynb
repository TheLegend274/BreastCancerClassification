```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets # This has basic data sets we can use
from sklearn.model_selection import train_test_split # We can train and test our data by importing this
```

Data Collection

```python
#loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset # 0 = B, 1 = M
```

```python
#Loading the data into a dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
data_frame.head()
```

```python
 #Making a new col caleld label, and the data will be used from the target values in the data set
 data_frame["target"] = breast_cancer_dataset.target
 data_frame
```

```python
data_frame.shape
```

```python
data_frame.info
```

```python
#checking how many missing values
data_frame.isnull().sum()
```

```python
#countuing how many B and M cases
data_frame['target'].value_counts()
```

1 -> Benign

2 -> Malignant

```python
data_frame.groupby('target').mean()
```

**Spliting the target and the features**

```python
x = data_frame.drop(columns = 'target', axis = 1)#input
y = data_frame['target']# output
```

```python
x
```

```python
y
```

**Splitting the data into training & testing data**

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2, random_state = 2 )
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

**Standardize the Data**

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
x_train_std = scalar.fit_transform(x_train)
x_test_std = scalar.transform(x_test)
```

**Building the Neural Network**

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
#Setting up layers
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)), #Input layer, 30  = number of features, flattens it into a one-dimensional array.
    keras.layers.Dense(20,activation = 'relu'), #Hidden Layer
    keras.layers.Dense(2, activation = 'sigmoid'), #Output Layer
])
```

```python
#compling neural network
model.compile(optimizer = 'adam', #The optimizer is responsible for updating the model's weights based on the loss function's gradient.
              loss = 'sparse_categorical_crossentropy', #loss function
              metrics = ['accuracy']) #Metrics are used to monitor the performance of the model during training and evaluation.
```

```python
#training the Neural network
history = model.fit(x_train_std, y_train, validation_split=0.1,epochs= 10)
```

**Visualizing accuracy and loss**

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['train','validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

**Accuracy of the model on test data**

```python
loss, accuracy = model.evaluate(x_test_std,y_test)
print(accuracy, loss)
```
